# Epitranscriptomics w/ Long-Read Sequencing Technologies

> **Introduction:** Overview of the epitranscriptomics field; the promise of native RNA sequencing for RNA modification detection.
>
> **Hands-on session overview:** Participants are split into groups of 5-6; each participant will analyze a separate dataset. Participants are encouraged to test different settings of the pipeline, discuss their results in their groups, and compare their findings.
>
> **Data and software:** As a case study, we will be working with tRNAs from a yeast KO strain; data that has already been basecalled and aligned will be provided. The AMaNITA tool will be used for downstream analyses and interpretations.

## 🍄 AMaNITA

AMaNITA (Abundances, Modifications, and Nanopore Intensity Toolbox/Application) was designed to perform statistical comparisons between groups of interest downstream of basecalling and alignment, and specifically caters to the analysis of Nano-tRNAseq data.



Take some time to go over the documentation; you can find the README and the wiki pages in the github folder.

### 👟 Run AMaNITA

⚠️ Due to discrepancies between the conda env and singularity, it is suggested that the commands are ran from the command line, after deactivating any conda env apart from base

```bash

#  AMaNITA help message
bash /home/train/longTREC/software/AMaNITA/src/AMaNITA.sh

# Set up your experiment
project_id=demo
wrk_dir=/home/train/longTREC/day3/output

bash /home/train/longTREC/software/AMaNITA/src/AMaNITA.sh \
--amanita_dir /home/train/longTREC/software/AMaNITA \
--project_id $project_id \
--wrk_dir $wrk_dir \
--data_dir ~/longTREC/day3/data/epitranscriptomics \
--organism "sacCer3" \
-01234 \
--mode_rnas "cyto" \
--singularity_dir /home/train/longTREC/software/singularity

# feel free to play with the command (mode_rnas and/or modules)

```

### 📑 Check the Report
